# Distributions in CUQIpy

This is a short demo showing how to define a random variable distribution in CUQIpy, and how to draw samples from it and visualize the results.

```python
import cuqi
import numpy as np
import matplotlib.pyplot as plt

# Define a 2D Gaussian distribution and draw samples from it
x = cuqi.distribution.Gaussian(mean=np.zeros(2), cov=1)
samples = x.sample(100)

# Visualize the probability density function
cuqi.utilities.plot_2D_density(x, -4, 4, -4, 4)

# Visualize the samples
samples.plot_pair(ax=plt.gca());
```
